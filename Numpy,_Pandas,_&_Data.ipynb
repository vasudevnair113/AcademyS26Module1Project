Let's get some more statistics on our tweet data now!!\
\
`:.3f`\
This f-string extension means rounded to 3 decimal places

```python
# Basic statistics
sentiment_scores = df['sentiment_score']
print("Sentiment Score Statistics (Rounded to 3 decimal places):")
print(f"Mean: {np.mean(sentiment_scores):.3f}")
print(f"Median: {np.median(sentiment_scores):.3f}")
print(f"Standard Deviation: {np.std(sentiment_scores):.3f}")
print(f"Min: {np.min(sentiment_scores):.3f}")
print(f"Max: {np.max(sentiment_scores):.3f}")

# Distribution of sentiment categories
print("Sentiment Distribution:")
sentiment_counts = df['sentiment_category'].value_counts() # Gets the number of unique values
print(sentiment_counts)

# Distribution of sentiment percentages
print("Sentiment Percentages:")
print((sentiment_counts / len(df) * 100).round(2))
```

# Data Visualization
We can READ the data but can we SEE it? 🤔🤔🤔

Pictures == 1000(words)

```python
# Set up the plotting style
plt.style.use('seaborn-v0_8') # Visual theme of our plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10)) # Creates plotting canvas
# fig = overall variable, axes = numpy array which holds the four subplots

# 1. Sentiment score distribution
axes[0, 0].hist(df['sentiment_score'], bins=10)
axes[0, 0].set_title('Distribution of Sentiment Scores')
axes[0, 0].set_xlabel("Sentiment Score")
axes[0, 0].set_ylabel("Frequency")
axes[0, 0].axvline(x=0, color='red', linestyle='--', label="Neutral")
axes[0, 0].legend()

# 2. Sentiment category counts
sentiment_counts.plot(kind='bar', ax=axes[0, 1], color=['red', 'gray', 'green'])
axes[0, 1].set_title('Sentiment Category Distribution')
axes[0, 1].set_xlabel("Sentiment Category")
axes[0, 1].set_ylabel("Count")

# 3. Sentiment vs Likes scatter plot
axes[1, 0].scatter(df['sentiment_score'], df['likes'])
axes[1, 0].set_title('Sentiment Score vs Likes')
axes[1, 0].set_xlabel("Sentiment Score")
axes[1, 0].set_ylabel("Likes")

# 4. Average engagement by sentiment
engagment_means = df.groupby('sentiment_category')[['likes', 'retweets']].mean()
axes[1, 1].set_title('Average Engagement by Snetiment')
axes[1, 1].set_xlabel('Sentiment Category')
axes[1, 1].set_ylabel('Average Count')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```
